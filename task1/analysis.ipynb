For each measurement point we have 
* What is measured,
* when

* value
* units
* metadata
* data origin

I assume it is reasonable to keep all the measurements in the one table to be able to serve wide range of queries, for example per location, per date, per measurement type.

Another approach would be to store the data for different measurement types in sdifferent tables. It would be reasonable we we know that 
* new types of measurements are rarely created 
* there are only few types of measurements

I assume the measurements are frequently accessed and metadata is rarely accessed and modified.
